# Étape n°2 : Nettoyage et préparation des données 


```python
# 1. Nettoyer les titres de colonnes :
# Supprimer les espaces au début et à la fin des titres (s’il y en a)

df_pop.columns = df_pop.columns.str.strip()
df_veg.columns = df_veg.columns.str.strip()
df_anim.columns = df_anim.columns.str.strip()
```

```python
# 4.Pour les datasets df_anim et df_veg, ajoutez une colonne ‘type’ qui prendra respectivement une valeur ‘animal’ et ‘vegetal’.
df_anim = df_anim.assign(type='animal')
df_veg = df_veg.assign(type='vegetal')
```

La différence entre les formats long et wide est la manière dont les données sont organisées. Dans le format long, chaque observation est représentée par une ligne, et chaque variable est représentée par une colonne. Dans le format wide, chaque observation est représentée par une colonne, et chaque variable est représentée par une ligne.

Dans notre cas, nous sommes en train de passer d'un format long à un format wide, car nous transformons les données pour que chaque élément soit représenté par une colonne.

```python
# 7. Créez une nouvelle colonne taux_croissance_pop_18_19 avec le taux de variation de la population entre 2018 et 2019 dans chaque pays. 

df_pop_pivot  = pd.pivot_table(df_pop, 
                                index=['area'],
                                columns='year',
                                values='pop_1000_hab')
df_pop_pivot = df_pop_pivot.reset_index()
df_pop_pivot['taux_croissance_pop_18_19'] = ((df_pop_pivot[2019] - df_pop_pivot[2018]) / df_pop_pivot[2018]) * 100
df_pop_pivot 
```

```python
#9. Calculez la disponibilité de nourriture totale par pays et par année, en kcal et kg de protéines. Attention aux unités de mesure !

df_food = df[['food_supply_(kcal/capita/day)', 'protein_supply_quantity_(g/capita/day)', 'area', 'year', 'pop_1000_hab','zone_geographique']]
df_food = df_food.groupby(['year', 'area','zone_geographique']).agg({'food_supply_(kcal/capita/day)': 'sum', 'protein_supply_quantity_(g/capita/day)': 'sum', 'pop_1000_hab': 'first'})
df_food_by_pop = df_food.copy()

def multiply_by_pop(row):
    row['food_supply_(kcal/capita/day)'] *= row['pop_1000_hab']*1000
    row['protein_supply_quantity_(g/capita/day)'] *= row['pop_1000_hab']*1000
    return row[['food_supply_(kcal/capita/day)', 'protein_supply_quantity_(g/capita/day)']] * 365


df_food_by_pop[['food_supply_(kcal/capita/day)', 'protein_supply_quantity_(g/capita/day)']] = df_food_by_pop.apply(multiply_by_pop, axis=1)
df_food_by_pop = df_food_by_pop.rename(columns={'food_supply_(kcal/capita/day)': 'food_supply_(kcal/pop/year)', 'protein_supply_quantity_(g/capita/day)': 'protein_supply_quantity_(g/pop/year)'})
df_food_by_pop
```

```python
# 10.Calculez le ratio énergie/poids de chaque produit et pays. Vous devriez vous apercevoir qu’étonnement, ces informations varient en fonction du pays. Pour pallier ce problème, calculez la moyenne de ce ratio pour chaque aliment.  Attention à bien gérer les valeurs égales à 0. Vérifiez la cohérence de votre calcul en comparant le résultat avec l’apport calorique d’un œuf.

df_energie = df[['item','food_supply_(kcal/capita/day)','food_supply_quantity_(kg/capita/yr)']].copy()
df_energie = df_energie[(df_energie['food_supply_(kcal/capita/day)'] >= 0.1) & (df_energie['food_supply_quantity_(kg/capita/yr)'] >= 0.1)]
df_energie['ratio_kcal_g'] = df_energie['food_supply_(kcal/capita/day)'] / ((df_energie['food_supply_quantity_(kg/capita/yr)'] /365 * 1000))
df_energie.replace([np.inf, -np.inf], np.nan, inplace=True)
df_energie.dropna(how='any',inplace=True)
df_energie=df_energie.groupby('item').mean()
df_energie
```

```python
# Vérifiez la cohérence de votre calcul en comparant le résultat avec l’apport calorique d’un œuf.

# Œuf entier : 140 kcal / 100 g

df_energie.loc['Eggs']

# Le ratio semble correct si on fait 100g par le ratio on tombe sur 137 kcal ce qui est plus ou moins égal à 140. 
```

```python
# 11. À l’instar de la question précédente, calculez le pourcentage de protéine de chaque aliment. Vérifiez votre résultat en le comparant avec l’apport en protéines d’un œuf.

df_prot = df[['item','protein_supply_quantity_(g/capita/day)','food_supply_quantity_(kg/capita/yr)']].copy()
df_prot = df_prot[(df_prot['protein_supply_quantity_(g/capita/day)'] >= 0.1) & (df_prot['food_supply_quantity_(kg/capita/yr)'] >= 0.1)]
df_prot['ratio_prot_g_(en %)'] = (df_prot['protein_supply_quantity_(g/capita/day)'] / ((df_prot['food_supply_quantity_(kg/capita/yr)'] /365 * 1000))*100)
df_prot.replace([np.inf, -np.inf], np.nan, inplace=True)
df_prot.dropna(how='any',inplace=True)
df_prot=df_prot.groupby('item').mean()
df_prot
```

```python
# Vérifiez votre résultat en le comparant avec l’apport en protéines d’un œuf.

# Pourcentage protéines pour 100g d'oeuf : 12,7 %.

df_prot.loc['Eggs']

# Le résultat est plus ou moins cohérent. 
```

```python
# 15. Avec la disponibilité alimentaire combien d’être humains pourrait-on nourrir ? Pour répondre à cette question vous avez besoin d’estimer les besoins alimentaires moyens d’une personne, effectuer une recherche pour trouver cette information. Interpréter vos résultats. Exprimer vos résultats sous la forme de % de la population mondiale.

# Selon mes recherches un être humain moyen a besoin d'environ 2200kcal par jour pour combler ses besoins énergétiques.

df_food_by_pop_2019 = df_food_by_pop.loc[2019]

disponibilite_alimentaire_par_jour = df_food_by_pop_2019['food_supply_(kcal/pop/year)'].sum()
nb_personne=disponibilite_alimentaire_par_jour/2200/365
pourcentage=(nb_personne/(population_2019*1000))*100

print(f"Nous pourrions nourrir {int(nb_personne)} personnes avec la disponibilité alimentaire. Soit {int(pourcentage)}% de la population en 2019.")
```

```python
# Et avec la disponibilité alimentaire en végétaux ?

df_vegetaux = df[(df['year'] == 2019) & (df['type'] == 'vegetal')][['food_supply_(kcal/capita/day)', 'area', 'year', 'pop_1000_hab']]
df_vegetaux = df_vegetaux.groupby('area').agg({'food_supply_(kcal/capita/day)': 'sum', 'pop_1000_hab': 'first'})
df_vegetaux_by_pop = df_vegetaux.copy()

def multiply_by_pop(row):
    row['food_supply_(kcal/capita/day)'] *= row['pop_1000_hab']*1000
    return row['food_supply_(kcal/capita/day)']

df_vegetaux_by_pop['food_supply_(kcal/capita/day)'] = df_vegetaux_by_pop.apply(multiply_by_pop, axis=1)
df_vegetaux_by_pop = df_vegetaux_by_pop.rename(columns={'food_supply_(kcal/capita/day)': 'food_supply_(kcal/pop/day)'})

disponibilite_alimentaire_par_jour = df_vegetaux_by_pop['food_supply_(kcal/pop/day)'].sum()
nb_personne=disponibilite_alimentaire_par_jour/2200
pourcentage=(nb_personne/(population_2019*1000))*100

print(f"Nous pourrions nourrir {int(nb_personne)} personnes avec la disponibilité alimentaire en végétaux. Soit {int(pourcentage)}% de la population en 2019.")
```
